```python
import folium
from pandas import DataFrame
ex = {'측정국' : ['풍무초교','김포공고','신월동주택','고강아파트','고강초교','신월코아루','고리올초','신월체육센터','신남초교','고척중교','길훈아파트','양도중교','신촌마을'
             ,'SK허브수','동화빌라','롯데낙천대'],
    '경도' : [126.726880,126.710673,126.832333,126.823488,126.819857,126.829531,126.827016,126.834902,126.844724,126.862002,126.850911,126.730661,126.752360,126.872651,126.860828,126.867474],
     '위도' : [37.605321,37.620515,37.535547,37.530633,37.527087,37.531672,37.526840,37.520089,37.515850,37.500462,37.484694,37.604599,37.595768,37.495873,37.477456,37.491960]
     }

ex=DataFrame(ex)
ex
```

```python
m=folium.Map(
location=[37.558492, 126.803263],zoom_start=12)


for i in ex.index:
    sub_lat =  ex.loc[i,'위도']
    sub_long = ex.loc[i,'경도']
    
    title = ex.loc[i,'측정국']
    
   
    folium.Marker([sub_lat,sub_long],tooltip = title).add_to(m)

m.save('example.html')
m
```
